# **Imports**


```python

from  matplotlib import pyplot as plt
import keras_ocr
import easyocr
```

# **Downloading the pretrained model**

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

# **Making Predictions using [Keras-OCR](https://keras-ocr.readthedocs.io/en/latest/)**



```python
#reading the images from their url
images = [
    keras_ocr.tools.read(url) for url in [
        '71FBiIeM2mS._SL1500_.jpg',
        'large.png',
        'MicrosoftTeams-image.jpg'
    ]
]

#making predictions
prediction_groups = pipeline.recognize(images)

# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(80, 60))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
  
    
```

**Making Prediction using [EasyOCR](https://github.com/JaidedAI/EasyOCR)**

```python
reader = easyocr.Reader(['en'])
results = [ reader.readtext(path, detail = 0, paragraph = True) for path in 
            [
               '71FBiIeM2mS._SL1500_.jpg',
               'large.png',
               'MicrosoftTeams-image.jpg'               
            ]
         ]

k = 1
for result in results:
    print("Image {}:\n".format(k))
    print('\n'.join(result))
    print("\n\n")
    k = k+1
```
